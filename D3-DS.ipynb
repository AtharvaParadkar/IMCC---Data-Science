#### Seperate Input and Output Variables

```python
x = df['YearsExperience'].values
y = df['Salary']
```

```python
x
```

```python
y
```

```python
x = x.reshape(30, 1)
x
```

#### Import the class

```python
from sklearn.linear_model import LinearRegression
```

#### Create the Object of Linear Regression

```python
regressor = LinearRegression()
```

#### Train the Algorithm with Data

```python
regressor.fit(x, y)
```

#### Predict on unseen data

```python
regressor.predict([[25]])
```
